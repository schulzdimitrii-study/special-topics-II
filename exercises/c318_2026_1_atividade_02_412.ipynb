## <font color='green'>**2.5 Identificação de Problemas nos Dados**</font>

### <font color='cyan'> Q16: Inspecione o conjuntos de dados para verificar se há:</font>
- #### <font color='cyan'> Dados Ausentes;</font>
- #### <font color='cyan'> Valores Atípicos (via IQR);</font>
- #### <font color='cyan'> Valores Duplicados;</font>
### <font color='cyan'> Reporte sua(s) descoberta(s) a seguir.</font>

```python
missing_values_count = X.isnull().sum()
print("------------------------------------")
print("Valores Ausentes por coluna:")
print(missing_values_count[missing_values_count > 0].sort_values(ascending=False).to_string())
print()
print(f"Total de valores ausentes: {missing_values_count.sum()}")
print(f"Porcentagem de valores ausentes: {missing_values_count.sum() / X.size:.2%}")
print("------------------------------------")

Q1 = X['capital-gain'].quantile(0.25)
Q3 = X['capital-gain'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = X[(X['capital-gain'] < lower_bound) | (X['capital-gain'] > upper_bound)]
print(f"Valores atípicos: {len(outliers)}")
print("------------------------------------")

duplicate_rows = X[X.duplicated()]
print(f"Linhas duplicadas: {len(duplicate_rows)}")
```

### <font color='cyan'> Q17: Caso um problema tenha sido encontrado, proponha uma solução para ele.</font>
### <font color='cyan'> Exiba o conjunto de dados antes e depois da solução, enfatizando as mudanças.</font>

```python
print("Antes da remoção de duplicatas:", X.shape[0])
X_cleaned = X.drop_duplicates()
print("Depois da remoção de duplicatas:", X_cleaned.shape[0])
print()
print("Antes da remoção de outliers:", X.shape[0])
X_no_outliers = X_cleaned[(X_cleaned['capital-gain'] >= lower_bound) & (X_cleaned['capital-gain'] <= upper_bound)]
print("Depois da remoção de outliers:", X_no_outliers.shape[0])
print()
print("Antes da remoção de valores ausentes:", X_no_outliers.shape[0])
X_no_missing = X_no_outliers.dropna()
print("Depois da remoção de valores ausentes:", X_no_missing.shape[0])
print("--------------------------------------")
print("Total de amostras antes e depois da remoção dos dados:")
print("Antes:", X.shape[0])
print("Depois:", X_no_missing.shape[0])
```

### <font color='cyan'> Q18: Escolha duas características numéricas do conjuntos de dados. Escreva seus nomes a seguir.</font>

age, capital-gain

### <font color='cyan'> Q19: Calcule e exiba o Coeficiente de Correlação de Pearson ($r$) entre elas.</font>

```python
pearson_r = X['age'].corr(X['capital-gain'])
print(f"Coeficiente de Correlação de Pearson (r): {pearson_r:.4f}")
```

### <font color='cyan'> Q20: Interprete o coeficiente de correlação, em termos de sinal e magnitude.</font>

A correlação positiva é muito fraca entre idade e capital-gain. O valor positivo sugere que há uma leve tendência de que, conforme a idade aumenta, o capital-gain também aumenta ligeiramente. Na prática, a idade isoladamente não é um bom preditor do ganho de capital, indicando que outros fatores são muito mais importantes para determinar o capital-gain de um indivíduo.
